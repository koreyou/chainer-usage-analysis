```python
import sentencepiece as spm
import re
from langdetect import detect
import fnmatch
import os
import glob
from collections import Counter
import codecs
import json

import numpy as np
import matplotlib.pyplot as plt
```

```python

spm.SentencePieceTrainer.Train('--input=function_names.txt --model_prefix=m --vocab_size=2000 --hard_vocab_limit=false')
```

```python
sp = spm.SentencePieceProcessor()
sp.Load("m.model")
```

```python
tokens = []
with open("function_names.txt") as fin:
    for line in fin:
        #print(sp.EncodeAsPieces(line.strip()))
        for t in sp.EncodeAsPieces(line.strip()):
            tokens.append(t.strip('\xe2\x96\x81'))
```

```python
tokens
```

```python
from wordcloud import WordCloud

text = " ".join(tokens)
# Generate a word cloud image
wordcloud = WordCloud(height=600, width=1000, background_color=None, mode="RGBA").generate(text)

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.close()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
wordcloud.to_file("function_name_wordcloud.png")
plt.show()
```
